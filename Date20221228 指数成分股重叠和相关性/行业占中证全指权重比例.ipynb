```python
from pandas import read_csv, read_excel
from os import getcwd, sep
from matplotlib import pyplot
import re
```

```python
def get_weight_from_zz(index_symbol: str):
    """
    中证指数

    :param index_symbol:
    :return:
    """
    weight = getcwd() + sep + r"data" + sep + r"{}closeweight.xls".format(index_symbol)
    weight = read_excel(io=weight, dtype=str)
    weight.columns = [''.join(re.findall(re.compile(u'[\u4e00-\u9fa5]'), i)) for i in weight.columns]
    weight["权重"] = weight["权重"].astype(float)
    weight.sort_values("权重", inplace=True, ascending=False)
    weight = weight.set_index("成分券代码")
    weight.index.name = "symbol"
    return weight


def get_weight_from_gz(file_name: str):
    """
    国证指数
    :param file_name:
    :return:
    """
    weight = getcwd() + sep + r"data" + sep + r"{}.xls".format(file_name)
    weight = read_excel(io=weight, dtype=str)
    weight.columns = [''.join(re.findall(re.compile(u'[\u4e00-\u9fa5]'), i)) for i in weight.columns]
    weight["权重"] = weight["权重"].astype(float)
    weight.sort_values("权重", inplace=True, ascending=False)
    weight = weight.set_index("样本代码")
    weight.index.name = "symbol"
    return weight
```

```python
weight = dict()
weight["中证全指"] = get_weight_from_zz("000985")

# 中证银行指数 (399986)
# 为反映中证全指指数样本中不同行业公司证券的整体表现，为投资者提供分析工具，将中证全指指数样本按中证行业分类分为11个一级行业、35个二级行业、90余个三级行业及200余个四级行业，再以进入各一、二、三、四级行业的全部证券作为样本编制指数，形成中证全指行业指数。
weight["中证银行指数"] = get_weight_from_zz("399986")

# 中证保险主题指数 (930618)
# 中证保险主题指数选取涉及保险、互联网保险业务或者参股保险企业的上市公司证券作为样本，以反映保险类相关上市公司的整体表现。
weight["中证保险主题指数"] = get_weight_from_zz("930618")

# 中证全指金融地产指数 (000992)
# 中证全指金融地产指数选取中证全指指数样本中属于金融和房地产两个中证一级行业的上市公司证券作为指数样本，以反映中证全指指数样本中金融和房地产上市公司证券的整体表现。
weight["中证全指金融地产指数"] = get_weight_from_zz("000992")

# 中证全指证券公司指数 (399975)
# 为反映中证全指指数样本中不同行业公司证券的整体表现，为投资者提供分析工具，将中证全指指数样本按中证行业分类分为11个一级行业、35个二级行业、90余个三级行业及200余个四级行业，再以进入各一、二、三、四级行业的全部证券作为样本编制指数，形成中证全指行业指数。
weight["中证全指证券公司指数"] = get_weight_from_zz("399975")

# 中证全指房地产指数 (931775)
# 为反映中证全指指数样本中不同行业公司证券的整体表现，为投资者提供分析工具，将中证全指指数样本按中证行业分类分为11个一级行业、35个二级行业、90余个三级行业及200余个四级行业，再以进入各一、二、三、四级行业的全部证券作为样本编制指数，形成中证全指行业指数。
weight["中证全指房地产指数"] = get_weight_from_zz("931775")

# 中证新能源汽车产业指数 (930997)
# 中证新能源汽车产业指数从沪深市场中选取业务涉及新能源汽车产业的上市公司证券作为指数样本，以反映新能源汽车产业上市公司证券的整体表现。
weight["中证新能源汽车产业指数"] = get_weight_from_zz("930997")

# 中证新能源指数 (399808)
# 中证新能源指数选取沪深市场中涉及可再生能源生产、新能源应用、新能源存储以及新能源交互设备等业务的上市公司证券作为指数样本，以反映新能源产业相关上市公司证券的整体表现。
weight["中证新能源指数"] = get_weight_from_zz("399808")
```
